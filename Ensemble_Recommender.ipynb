```python
import pandas as pd
import numpy as np
from numpy import *
```

```python
model1 = pd.read_csv('factorization.csv')
model2 = pd.read_csv('RandomForest.csv')
model3 = pd.read_csv('gradient_boosted_tree.csv')
model4 = pd.read_csv('decisionTree.csv')
model5 = pd.read_csv('SVM.csv')
model6 = pd.read_csv('logisticRegression.csv')
model7 = pd.read_csv('prediction_mean.csv')
```

```python
def sortResults (modelResults):

  modelResults['UserID'] = modelResults['TrackID'].str.split('_').str[0].astype(int)
  modelResults['TrackID'] = modelResults['TrackID'].str.split('_').str[1].astype(int)
  modelResults = modelResults.sort_values(by=['UserID','TrackID'])
  return modelResults
```

```python
model1 = sortResults(model1)
model2 = sortResults(model2)
model3 = sortResults(model3)
model4 = sortResults(model4)
model5 = sortResults(model5)
model6 = sortResults(model6)
model7 = sortResults(model7)
```

```python
model1.Predictor.replace(0,-1,inplace= True)
model2.Predictor.replace(0,-1,inplace= True)
model3.Predictor.replace(0,-1,inplace= True)
model4.Predictor.replace(0,-1,inplace= True)
model5.Predictor.replace(0,-1,inplace= True)
model6.Predictor.replace(0,-1,inplace= True)
model7.Predictor.replace(0,-1,inplace= True)
```

```python
s1 = model1.Predictor.tolist()
s2 = model2.Predictor.tolist()
s3 = model3.Predictor.tolist()
s4 = model4.Predictor.tolist()
s5 = model5.Predictor.tolist()
s6 = model6.Predictor.tolist()
s7 = model7.Predictor.tolist()



S=np.transpose(mat([s1,s2,s3,s4,s6,s7])) #customize here
print(S)
```

```python
score1 = 0.85567
score2 = 0.85575
score3 = 0.85564
score4 = 0.85567
score5 = 0.84660
score6 = 0.85526
score7 = 0.84797

score_mat=np.transpose(mat([score1,score2,score3,score4,score6, score7])) #customize here
k=120000 #number of samples
STx=k*(2*score_mat-1)

print(STx)
```

```python
STS=np.transpose(S)*S
STS_inverse=np.linalg.pinv(STS)

a=STS_inverse*STx
S_new=S*a
S_new=S_new.T
new_S=np.array(S_new).reshape((20000,6)).tolist()
print(new_S)
```

```python
def prediction_table(grades_table):
    prediction_table = [0 for i in range(len(grades_table)*6)]
    for i in range(len(grades_table)):
        x = np.array(grades_table[i])
        x = x.argsort()

        for k in range(6):
            if k<3:
                prediction_table[i * 6 + x[k]] = 0
            else:
                prediction_table[i * 6 + x[k]] = 1
      
    return prediction_table
```

```python
new_ratings_table = prediction_table(new_S)
```

```python
model2.Predictor = new_ratings_table #overwrite predictions
```
